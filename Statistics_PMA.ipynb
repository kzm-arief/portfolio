```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
%matplotlib inline
```

```python
df = pd.read_excel("data_beasiswa.xlsx")
df.head()
```

```python
df.info()
```

```python
features = ["IPK", "Skor", "Listening", "Reading", "Speaking", "Writing"]
x = df.loc[:, features].values
x
```

```python
y = df.loc[:,["Home University"]].values
y
```

```python
z = df.loc[:,["Host University"]].values
z
```

```python
x = StandardScaler().fit_transform(x)
pd.DataFrame(data = x, columns = features).head()
```

```python
pca = PCA(n_components=4)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ["PC_1", "PC_2", "PC_3", "PC_4"])
principalDf.head(10)
```

```python
df[["Home University"]].head(10)
```

```python
df[["Host University"]].head(10)
```

```python
finalDf1 = pd.concat([principalDf, df[["Home University"]]], axis = 1)
finalDf1.head(10)
```

```python
finalDf2 = pd.concat([principalDf, df[["Host University"]]], axis = 1)
finalDf2.head(10)
```
